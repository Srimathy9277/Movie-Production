### Clarifying how are popularities of movies related with the movie budgets:

```python
correlation = production['popularity'].corr(production['budget'])

print(f"Correlation between Popularity and Budget: {correlation:.2f}")
```

### Ckecking whether are they correlated or totally uncorrelated with each other:

Interpretation of the correlation coefficient:

If the correlation coefficient (often denoted as "r") is close to 1; it indicates a strong positive correlation, meaning that as movie budgets increase, popularity tends to increase as well.

If the correlation coefficient is close to -1; it indicates a strong negative correlation, meaning that as movie budgets increase, popularity tends to decrease.

If the correlation coefficient is close to 0; it suggests little to no linear correlation between the two variables, implying that movie budgets and popularity are not strongly related.



The value of the correlation coefficient will help you interpret the relationship. A positive value indicates a positive correlation, while a negative value suggests a negative correlation.

### Sorting the data in descending order based on revenue and filter the top 500 movies

### Calculating the measures of central tendency for the specified columns

```python
mean_budget = top_500_movies['budget'].mean()
median_budget = top_500_movies['budget'].median()
mean_revenue = top_500_movies['revenue'].mean()
median_revenue = top_500_movies['revenue'].median()
mean_runtime = top_500_movies['runtime'].mean()
median_runtime = top_500_movies['runtime'].median()
```

```python
print(f"Measures of Central Tendency (Top 500 Movies):")
print(f"Budget - Mean: {mean_budget:.2f}, Median: {median_budget:.2f}")
print(f"Revenue - Mean: {mean_revenue:.2f}, Median: {median_revenue:.2f}")
print(f"Runtime - Mean: {mean_runtime:.2f}, Median: {median_runtime:.2f}")
```
